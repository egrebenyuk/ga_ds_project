## Load open source libraries

```python
import pandas as pd
import numpy as np
```

## Define useful functions

```python
def get_feed_info(df):
    # print general info about data frame
    print "Shape of dataset: ", df.shape
    
    # check on missing values
    rows_count = len(df)
    rows_with_missing_data = df.shape[0] - df.dropna().shape[0]
    print "There are ", rows_with_missing_data, " row(s) out of ", rows_count, " have missing values."

def get_target_stats(df):
    print "\nCounts by satisfaction:"
    all = float(len(df))
    satisfied = df[df.TARGET==0]['ID'].count()
    unsatisfied = df[df.TARGET==1]['ID'].count()
    print "Satisfied customers:", satisfied, "(", round(satisfied*100/all), "% )"
    print "Unsatisfied customers:", unsatisfied, "(", round(unsatisfied*100/all), "% )"
```

## Load datasets

```python
df_train = pd.read_csv("data/train.csv")
get_feed_info(df_train)
get_target_stats(df_train)
```

```python
df_test = pd.read_csv("data/test.csv")
get_feed_info(df_test)
```

## Descriptive statistics

```python
d_stats = df_train.describe()
pd.options.display.max_columns = d_stats.shape[1]
d_stats
```

The column of var3 contains error value of -999999. Also there are several columns (ind_var2_0, ind_var2) in this dataset have 0 value as min and max. I am going to remove those columns as far as they don't contain any useful information. Another words, they are empty.

## Check for empty columns

```python
# get empty columns
zero_columns = df_train.columns[df_train.iloc[:,:371].max() == 0].tolist()
print "The following", len(zero_columns), "columns out of", df_train.shape[1], "are empty:\n", zero_columns
```

```python
# Remove empty columns from both datasets
df_train.drop(zero_columns, axis=1, inplace=True)
print "Shape of train dataset after removing empty columns:", df_train.shape

df_test.drop(zero_columns, axis=1, inplace=True)
print "Shape of test dataset after removing empty columns:", df_test.shape
```

## Check for duplicated columns

```python
# find duplicated columns
dup_columns = []
columns = df_train.columns
for i in range(len(columns)-1):
    v = df_train[columns[i]].values
    for j in range(i+1,len(columns)):
        if np.array_equal(v,df_train[columns[j]].values):
            dup_columns.append(columns[j])

print "The following", len(dup_columns), "columns are duplicated:\n", dup_columns
```

```python
# Remove duplicated columns from both datasets
df_train.drop(dup_columns, axis=1, inplace=True)
print "Shape of train dataset after removing duplicated columns:", df_train.shape

df_test.drop(dup_columns, axis=1, inplace=True)
print "Shape of test dataset after removing duplicated columns:", df_test.shape
```

## var3 research

```python
df_train.var3.value_counts().head()
```

```python
# Set most popular value for var3 in case of error value of -999999 in both datasets
print "Error values in train dataset before update: ", df_train.loc[df_train.var3==-999999].shape
df_train = df_train.replace(-999999,2)
print "Error values in train dataset after update: ", df_train.loc[df_train.var3==-999999].shape

print "Error values in test dataset before update: ", df_test.loc[df_test.var3==-999999].shape
df_test = df_test.replace(-999999,2)
print "Error values in test dataset after update: ", df_test.loc[df_test.var3==-999999].shape
```

## Save clean datasets for further work

```python
# train dataset
df_train.to_csv('data/clean_train.csv', sep=',', index=False)

# test dataset
df_test.to_csv('data/clean_test.csv', sep=',', index=False)
```
